```python
import numpy as np
import matplotlib.pyplot as plt 
no_samples = 1000                               
x = np.linspace(0,2*np.pi, no_samples)         
y_sin = np.sin(x)
y_cos = np.cos(x)                               
plt.figure()                                     
plt.plot(x,y_sin, label='sin(x)')                     
plt.plot(x,y_cos, label='cos(x)')                     
plt.xlabel('x')
plt.ylabel('sin and cos(x)')
plt.title('plot of sine and cosine')
plt.grid()
plt.legend()   
```
